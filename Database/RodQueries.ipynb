#### I put a load method in rod_database.py has for unpickling the database for raw actions and another for creating a dataframe for pandas

```python
# the unpickling is really fast
items = rdb.load_db()
```

```python
df = rdb.make_pandas_df(items,'coerce')
```

```python
standard_cols=['ITEM_NAME', 'LEVEL','VALUE', 'AFFECTS','AREA','WORN','MOB','GENRES_ALLOWED','RACES_ALLOWED','HIT_ROLL','DAMAGE_ROLL']
weapon_cols=['ITEM_NAME', 'AREA', 'LEVEL', 'WEAPON_TYPE','AVERAGE_DAMAGE', 'VALUE', 'STR', 'INT', 'WIS', 'DEX', 'CON',
       'LCK', 'HP', 'MANA', 'HIT_ROLL', 'DAMAGE_ROLL']
```

```python
df.columns
```

#### Look for high damage weapons

```python
df_dmg=df.query('LEVEL<15 & AVERAGE_DAMAGE>10 and VALUE>3').sort_values(by='VALUE').iloc[:, : 20].fillna('')[weapon_cols]
df_dmg
```

#### Stabbing weapons!

```python
df_stab=df.query('LEVEL<30 & LEVEL >5 & VALUE >1 & WEAPON_TYPE.str.startswith("stab")').fillna('').sort_values(by='VALUE')[weapon_cols]
df_stab
```

#### Regeneration items we can use

```python
cols=['ITEM_NAME', 'LEVEL','VALUE', 'AFFECTS','AREA','MOB','HIT_ROLL','DAMAGE_ROLL','RACES_ALLOWED']
df_regen=df.query('LEVEL<35 & VALUE >1 & AFFECTS.str.find("REG")>-1').fillna('').sort_values(by='VALUE')[cols]
df_regen
```
